```python
from google import genai

client = genai.Client(api_key="AIzaSyDT2DU36k98zMC2OGSEjA46-y-8EChNqV0")

print("AI Email Assistant")
print("1 - Job Application Email")
print("2 - Leave Request Email")
print("3 - Meeting Request Email")

choice = input("Choose email type: ")

name = input("Enter your name: ")

print("\nChoose email tone:")
print("1 - Formal")
print("2 - Confident")
print("3 - Friendly")

tone_choice = input("Enter tone: ")

if tone_choice == "1":
    tone = "Formal and polite tone with a professional closing like 'Sincerely'"
elif tone_choice == "2":
    tone = "Confident and assertive tone with a closing like 'Best regards'"
else:
    tone = "Friendly and warm tone with a closing like 'Warm regards'"

# Email type prompts
if choice == "1":
    role = input("Enter job role: ")
    skills = input("Enter your skills: ")

    prompt = f"""
    Write a job application email.

    Applicant Name: {name}
    Role: {role}
    Skills: {skills}

    Tone: {tone}

    Limit: 120 words
    """

elif choice == "2":
    reason = input("Reason for leave: ")
    days = input("Number of days: ")

    prompt = f"""
    Write a professional leave request email.

    Employee Name: {name}
    Reason: {reason}
    Duration: {days} days

    Tone: {tone}

    Limit: 100 words
    """

else:
    topic = input("Meeting topic: ")

    prompt = f"""
    Write a professional meeting request email.

    Sender Name: {name}
    Meeting Topic: {topic}

    Tone: {tone}

    Limit: 100 words
    """

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print("\nGenerated Email:\n")
print(response.text)

#save email to a file
with open("Generated_email.text","w") as file:
  file.write(response.text)

print("\nEmail saved as generated_email.txt")
```
